```python
import requests
import numpy as np
from PIL import Image
```

```python

# URL of the FastAPI endpoint
url = 'http://localhost:80/predict'

# Path to the image file you want to test
example1 = 'example.jpeg'
example2 = 'example2.png'
```

```python
def get_model_prediction(url, image_path):
    img = Image.open(image_path).convert('L').resize((28, 28))
    image_array = np.array(img)
    image_array = image_array.reshape((1,1,28,28))
    with open(image_path, 'rb') as image_file:
        # Create a dictionary with the file
        files = {'data': image_array.tolist()}
        
        # Send the POST request
        response = requests.post(url, json=files)
    return response.json()
```

```python
Image.open(example1)
```

```python
get_model_prediction(url, example1)
```
